```python
import pandas as pd
```

```python
data = pd.read_csv("Airline_Delay_Cause.csv")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.drop(['carrier','carrier_name',"airport","airport_name"], axis=1 , inplace=True )
```

```python
data.info()
```

```python
data
```

```python
X = data.drop(['late_aircraft_ct'],axis = 1, inplace = False)
y = data['late_aircraft_ct']
```

```python
data.head()
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=44, shuffle =True)
```

```python
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

```python
X.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LinearRegressionModel = LinearRegression(fit_intercept=True,copy_X=True,n_jobs=-1)
```

```python
%%time
LinearRegressionModel.fit(X_train, y_train)
```

```python
print('Linear Regression Train Score is : ' , LinearRegressionModel.score(X_train, y_train))
print('Linear Regression Test Score is : ' , LinearRegressionModel.score(X_test, y_test))
print('Linear Regression Coef is : ' , LinearRegressionModel.coef_)
print('Linear Regression intercept is : ' , LinearRegressionModel.intercept_)
print('----------------------------------------------------')
```

```python
y_pred = LinearRegressionModel.predict(X_test)
print('Predicted Value for Linear Regression is : ' , y_pred[:10])
```

```python
from sklearn.metrics import accuracy_score
```

```python
list(y_pred)
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error 
```

```python
MSEValue = mean_squared_error(y_test, y_pred, multioutput='uniform_average')
print('Mean Squared Error Value is : ', MSEValue)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RandomForestRegressorModel = RandomForestRegressor(n_estimators=100,max_depth=2, random_state=33 ,n_jobs=-1)
```

```python
RandomForestRegressorModel.fit(X_train, y_train)
```

```python
print('Random Forest Regressor Train Score is : ' , RandomForestRegressorModel.score(X_train, y_train))
print('Random Forest Regressor Test Score is : ' , RandomForestRegressorModel.score(X_test, y_test))
print('----------------------------------------------------')
```

```python
y_pred = RandomForestRegressorModel.predict(X_test)
print('Predicted Value for Random Forest Regressor is : ' , y_pred[:10])
```

```python
list(y_pred)
```

```python
from sklearn.metrics import mean_absolute_error 
MAEValue = mean_absolute_error(y_test, y_pred, multioutput='uniform_average') 
print('Mean Absolute Error Value is : ', MAEValue)
```
